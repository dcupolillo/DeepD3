# Evaluate training

Have a look at Loss, Accuracy and Intersection-over-Union (IoU) across epochs for spines.

## Imports

```python
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Get curves from model logger.

```python
logger_path = Path(r"../data/models/model_251215_logger.csv")

history_df = pd.read_csv(logger_path)

epochs = history_df['epoch']

train_loss = history_df['spines_loss']
valid_loss = history_df['val_spines_loss']

train_iou = history_df['spines_iou_score']
valid_iou = history_df['val_spines_iou_score']

train_acc = history_df['spines_acc']
valid_acc = history_df['val_spines_acc']
max_acc = np.max(valid_acc)
```

## Loss plot

```python
fig_loss, ax_loss = plt.subplots(figsize=(8, 6))

ax_loss.plot(epochs, train_loss, label='Training', color="crimson")
ax_loss.plot(epochs, valid_loss, label='Validation', color="mediumblue")

ax_loss.set(
    xlabel='Epochs',
    ylabel='Loss',
    xlim=(0, epochs.max() + 1),
)
ax_loss.legend(loc="best", frameon=False)
```

## Accuracy plot

```python
fig_loss, ax_loss = plt.subplots(figsize=(8, 6))

ax_loss.plot(epochs, train_acc, label='Training', color="crimson")
ax_loss.plot(epochs, valid_acc, label='Validation', color="mediumblue")

ax_loss.set(
    xlabel='Epochs',
    ylabel='Accuracy',
    ylim=(0, 1),
    xlim=(0, epochs.max() + 1),
)
ax_loss.legend(loc="best", frameon=False)
```

## IoU plot

```python
fig_loss, ax_loss = plt.subplots(figsize=(8, 6))

ax_loss.plot(epochs, train_iou, label='Training', color="crimson")
ax_loss.plot(epochs, valid_iou, label='Validation', color="mediumblue")

ax_loss.set(
    xlabel='Epochs',
    ylabel='IoU',
    ylim=(0, 1),
    xlim=(0, epochs.max() + 1),
)
ax_loss.legend(loc="best", frameon=False)
```
